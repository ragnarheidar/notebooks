```python
import requests
import json
import pandas as pd
import os
```

```python
# sting to get volcanoes and their colors
# https://api.epos-iceland.is/v1/volcano/products/aviation_color_codes?bbox=62%2C-25%2C67%2C-11&format_type=JSON
# docs:
# https://docs.epos-iceland.is/#/Volcano%20obervations%20%3E%20Products%2C%20reports%20%26%20information/volcanoes_callaviationcolorcodes


# build request sting
base_url = 'https://api.epos-iceland.is/v1/volcano/products/aviation_color_codes?'
bounding_box = "bbox=62,-25,67,-11"
file_format = "format_type=JSON"

string = base_url + bounding_box + '&' + file_format
print(string)
```

```python
r = requests.get(string)
r.encoding = 'utf-8'
r.json()
```

```python
data = r.json()
with open(os.path.join('data/','VolcanoColors.json'), 'w') as f:
    json.dump(data, f)
f.close()
```
