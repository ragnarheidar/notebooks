```python
import rasterio
import lidario
from matplotlib import pyplot
```

```python
src_raster = rasterio.open('data/dem_sample_50x50_epsg3057.tif')
pyplot.imshow(src_raster.read(1), cmap='gray')
pyplot.show()
```

```python
from rasterio.plot import show
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
show(src_raster.read(1), transform=src_raster.transform, cmap='gray')
```

```python
raster_translator = lidario.Translator('geotiff', 'csv')
raster_translator.translate('data/dem_sample_50x50_epsg3057.tif', out_file='data/dem_points', band=1)
```
