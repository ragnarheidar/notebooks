### An example on how to use the Icelandic Met Office API for earthquake data

This example use The Skjálftalísa API avilable at [api.vedur.is](https://api.vedur.is/?urls.primaryName=Skj%C3%A1lftal%C3%ADsa)


```python
import os
import requests
# import json
# import pandas as pd
```

```python
# Let's start by building our reqest

# This is the basic URL for using ther quakefilter option
url = "https://api.vedur.is/skjalftalisa/v1/quakefilter"


headers = {"Content-Type": "application/json", "Accept": "application/json"}
# headers = {"Content-Type": "application/json", "Accept": "text/csv"}

json_params = {
  "area": [
    [64.1, -22.8],
    [64.1, -21.7],
    [63.8, -21.7],
    [63.1, -22.8]
  ],
  "depth_max": 25,
  "depth_min": 5,
  "end_time": "2023-10-27 00:00:00",
  "event_type": [
    "qu"
  ],
  "magnitude_preference": [
    "Mlw"
  ],
  "originating_system": [
    "SIL picks"
  ],
  "size_max": 6,
  "size_min": 2,
  "start_time": "2023-10-24 00:00:00"
}
```

```python
# adding things togeather and send a request with all our parameters
response = requests.post(url, headers=headers, json=json_params)

# print the status code of the API and outout the data we got back
# print("Status Code", response.status_code)
response.json()
```

```python
# save respnse as JSON file
data = response.json()
with open(os.path.join('data/','Earthquakes.json'), 'w') as f:
    json.dump(data, f)
f.close()
```
